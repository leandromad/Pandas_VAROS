# Link com o código dos países

https://en.wikipedia.org/wiki/List_of_ISO_3166_country_codes

```python
pib_percapta = wb.download(indicator='NY.GDP.PCAP.KD',
                    country=['BRA','USA','ARG','MEX'],
                    start=1980, end=2022)

print(pib_percapta)
```

```python
aumento_em_40_anos = pib_percapta.groupby('country').first()/pib_percapta.groupby('country').last() - 1

aumento_em_40_anos
```

```python
pib_percapta.loc['Brazil'].sort_index().plot()
```

# Exercícios

- Exercício 63: Busque os dados no banco mundial da evolução da emissão de carbono no mundo com o nome "CO2 emissions (kt)" desde 1990. Obs: Para puxar dados agregados do mundo a sigla a ser colocador é '1w'

```python
emissao_carbono = wb.download(indicator = 'EN.ATM.CO2E.KT',
                             country = '1w',
                             start = 1990, end = 2023)

emissao_carbono

emissao_carbono = emissao_carbono.dropna()
emissao_carbono = emissao_carbono.droplevel(level = 0)
emissao_carbono = emissao_carbono.sort_index()
emissao_carbono

```

```python
emissao_carbono.plot()
```
