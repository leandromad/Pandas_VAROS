```python
df = yf.download(['WEGE3.SA','VALE3.SA','PETR4.SA','LREN3.SA','PETR3.SA','^BVSP'], '2011-08-31','2022-08-31')['Adj Close']
retornos = df.pct_change().dropna()
retornos_ibov = retornos['^BVSP']
retornos = retornos.drop('^BVSP', axis = 1)
corr_movel = retornos.rolling(252).corr(retornos_ibov)
corr_movel = corr_movel.dropna()
corr_movel
```

```python
corr_movel.plot()
```

```python
corr_weg_vale = retornos['WEGE3.SA'].rolling(252).corr(retornos['VALE3.SA']).dropna().plot()
```
