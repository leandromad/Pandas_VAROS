```python
### Merge x Join


dados_roe = pd.DataFrame({
    'ticker': ['WEGE3','VALE3','PCAR3','ELET3','MGLU3'],
    'ROE': [20,9,19,3,5]
})

dados_roic = pd.DataFrame({
    'codigo':['WEGE3', 'VALE3','PCAR3','ELET3','VIIA3'],
    'ROIC': [25,6,13,1,2]
})

indicadores = pd.merge(dados_roe, dados_roic, how = 'inner',
                      left_on = ['ticker'], right_on = ['codigo'])

indicadores
```

```python
dados_cotacoes = pdr.get_data_yahoo('ABEV3.SA', start='2017-01-01')
dados_pl = pd.DataFrame({'ticker': ['ABEV3'] * len(dados_cotacoes),
                        'PL': np.random.randint(10, 20, len(dados_cotacoes))},
                       index = dados_cotacoes.index)

print(dados_pl)
print(dados_cotacoes)
```

- Exercício 77: Puxe, separadamente, dados de fechamento e volume das ações MGLU3.SA, CASH3.SA e AGRO3.SA. A partir dessas tabelas, utilize o melt para transformar as colunas com os tickers na variável 'empresa' e utilize essa chave 'empresa' para dar um merge entre os DataFrames.

```python
# Exercício 76

dados_mglu3 = pdr.get_data_yahoo('MGLU3.SA', start = '2017-01-01')
dados_viia3 = pdr.get_data_yahoo('VIIA3.SA', start = '2017-01-01')

dados_mglu3['empresa'] = ['MGLU3'] * len(dados_mglu3)
dados_viia3['empresa'] = ['VIIA3'] * len(dados_viia3)

dados_finais = pd.concat([dados_mglu3, dados_viia3], ignore_index = True)
dados_finais
```

```python
dados_close = pdr.get_data_yahoo(['MGLU3.SA','CASH3.SA','AGRO3.SA'], start = '2017-01-01')['Adj Close']
dados_close = pd.melt(dados_close, var_name = 'empresa', value_name = 'price')

dados_volume = pdr.get_data_yahoo(['MGLU3.SA','CASH3.SA','AGRO3.SA'], start = '2017-01-01')['Volume']
dados_volume = pd.melt(dados_volume, var_name = 'empresa', value_name = 'volume')

dados_finais = pd.merge(dados_close, dados_volume, on = 'empresa')
dados_finais
```
