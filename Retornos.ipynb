```python
import pandas as pd
import yfinance as yf
```

```python
df = yf.download('WEGE3.SA', start = '2017-01-01', end = '2022-08-31')[['Adj Close']]
df
```

```python
# retorno diário

retorno_diario = df.pct_change().dropna()
retorno_diario 
```

```python
# retorno anual
retorno_anual = df.resample('Y').last().pct_change().dropna()
retorno_anual
```

```python
# janela de retorno de 36 meses

dias = 252 * 3

dados_36m = df.pct_change(periods = dias).dropna()

dados_36m
```

```python
dados_36m.plot()
```

```python
dados_mercado = yf.download(['^BVSP','WEGE3.SA'], start = '2017-01-01', end = '2022-08-31')[['Adj Close']]
dados_mercado
```

```python
dados_36m = dados_mercado.pct_change(periods = dias).dropna()
dados_36m
```

```python
dados_36m.plot()
```
