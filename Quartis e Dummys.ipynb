```python
import pandas as pd
from pandas_datareader import data as pdr
import numpy as np
```

```python
valores = np.random.rand(10)

valores
```

```python
#dividindo valores baseado em %

divisoes = [0, 0.2, 0.4, 0.6, 0.8, 1]

valores_categorias = pd.cut(valores, divisoes)

valores_categorias
```

```python
numero_de_amostras_por_categoria = pd.value_counts(valores_categorias)

numero_de_amostras_por_categoria
```

```python
# transformar a amostra em vari√°veis dummys

df_dummies = pd.get_dummies(valores_categorias)
df_dummies
```

```python
# analise de idade de CEOS das empresas da bolsa

df_ceos_empresas = pd.DataFrame({
    'Empresa': ['Weg','Vale','Petrobras','Renner',
               'Assai', 'Nubank', 'Inter','BTG'],
    'Nome_ceo': ['Pedro', 'Nelson', 'Lucas','Stefany',
                'Mateus','Aline','Raquel','Paloma'],
    'Idade': [36,45,55,60,62,23,31,32]
})

df_ceos_empresas
```

```python
faixa_etaria = [0, 18, 25, 35, 50, 65]

ceos_por_idade = pd.value_counts(pd.cut(df_ceos_empresas['Idade'], faixa_etaria))

ceos_por_idade.sort_values()
```

```python
df = pd.read_excel('dados/Mundo20dy_yield_empresas_26set.xlsx')
df = df.dropna()
df
```

```python
df['ranking_dividendos'] = df['DY_12M'].rank(ascending = False)
df
```

```python
divisorias = [0, (max(df['ranking_dividendos'])/4),
             (max(df['ranking_dividendos'])/4 * 2),
             (max(df['ranking_dividendos'])/4 * 3),
             max(df['ranking_dividendos']) + 1]

divisorias
```

```python
df['quartil'] =  pd.cut(df['ranking_dividendos'], divisorias)

df['rank_quartil'] = df['quartil'].rank(method = 'dense')
df
```

```python
empresa_escolhida = input('''Qual empresa voc√™ quer saber o quartil que ela ocupa na distribui√ß√£o de dividendos da bolsa?''').upper()

quartil = (df[df['TICKER'] == empresa_escolhida]).iloc[0, 5]

print(f'''A empresa {empresa_escolhida} fica localizada no {quartil}¬∫ quartil em distribui√ß√£o de dividendos
de todas as empresas da bolsa.''')
```
